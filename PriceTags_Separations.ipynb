```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("./data/PriceTagsList.jpg")

cv2.imshow('Price Tag List',image)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.title('Price Tag List')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow('Grayscale Price Tags List',gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.title("Grayscale Price Tags List")
plt.imshow(cv2.cvtColor(gray_image,cv2.COLOR_BGR2RGB))
```

```python

_,thresh = cv2.threshold(gray_image,120,255,cv2.THRESH_BINARY_INV)

cv2.imshow('Threshold_GrayScale Price Tag List',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.title('Threshold_GrayScale Price Tag List')
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB))
```

```python
contours, _ = cv2.findContours(thresh,cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

filtered_contours = list(filter(lambda ctr: cv2.contourArea(ctr)>3000,contours))
```

```python
img_original= image.copy()
price_tags=[]
```

```python
for idx, cnt in enumerate(filtered_contours):
    x,y,w,h = cv2.boundingRect(cnt)

    # x=max(0,x-8)
    # y=max(0,y-15)
    # w= w +(2*8)
    # h= h+ 7+15

    if x+w > image.shape[1]:
        w=image.shape[1]-x


    cv2.rectangle(img_original,cv2.COLOR_BGR2RGB)

    price_tags.append((x,y,w,h))
    

```
